```python
import pandas as pd
```

```python
two_thousands_df = pd.read_csv("dataset-of-00s.csv")
two_thousands_df
```

```python
two_thousand_tens_df = pd.read_csv("dataset-of-10s.csv")
two_thousand_tens_df
```

```python
sixties_df = pd.read_csv("dataset-of-60s.csv")
sixties_df
```

```python
seventies_df = pd.read_csv("dataset-of-70s.csv")
seventies_df
```

```python
eighties_df = pd.read_csv("dataset-of-80s.csv")
eighties_df
```

```python
nineties_df = pd.read_csv("dataset-of-90s.csv")
nineties_df
```

```python
frames = [two_thousands_df, two_thousand_tens_df, sixties_df, seventies_df, eighties_df, nineties_df]
df = pd.concat(frames)
df
```

```python
cleaned_df = df.drop(['key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'duration_ms', 'time_signature', 'chorus_hit', 'sections', 'target'], axis = 1)
cleaned_df
```

```python
cleaned_df.to_csv('cleaned_spotify_data.csv', index=False)
```
